```python
# Diccionario para almacenar los resultados
resultados = {}

# Columnas principales a analizar
columnas = [f'C{i}' for i in range(1, 24)]  # C1 a C23

# Recorremos todas las columnas
for columna in columnas:
    if columna in df.columns:
        # Buscar subgrupos relacionados con esta columna
        subgrupos = [col for col in df.columns if col.startswith(f'subgrupo_{columna}')]

        for subgrupo in subgrupos:
            # Filtrar solo filas donde la variable principal es 0 o 1
            filas_validas = df[(df[columna] == 0) | (df[columna] == 1)]

            # Contar los 1 en el subgrupo
            cuenta_1 = (filas_validas[subgrupo] == 1).sum()

            # Guardar el resultado (una única vez por subgrupo)
            resultados[subgrupo] = cuenta_1

# Convertimos directamente el diccionario a un DataFrame
resultados_df = pd.DataFrame(list(resultados.items()), columns=['Subgrupo', 'Penalizaciones'])


# Mostramos el DataFrame
print(resultados_df)
```

```python
# Aseguramos que la columna con los subgrupos está correctamente identificada
columna_codigos = resultado_final.columns[0]

# Creamos un diccionario para saber a qué grupo pertenece cada subgrupo
grupo_por_codigo = {
    "C1": "C1. TIEMPO DE RESPUESTA",
    "C2": "C2. ACCEDER AL SERVICIO DESEADO",
    "C3": "C3. CONTACTAR/HABLAR CON UN ASESOR",
    "C4": "C4. TIEMPO DE TRANSFERENCIA",
    "C5": "C5. LLAMADA EN ESPERA (TÚNEL)",
    "C6": "C6. TOMA DE CONTACTO",
    "C7": "C7. IDENTIFICACIÓN",
    "C8": "C8. PERSONALIZACIÓN",
    "C9": "C9. MOTIVO DE CONTACTO",
    "C10": "C10. REFORMULACIÓN",
    "C11": "C11. DIRECCIÓN DE LA LLAMADA",
    "C12": "C12. TRANSFERENCIA SIN FRICCIÓN",
    "C13": "C13. SABER ORIENTAR",
    "C14": "C14. SABER ESCUCHAR/ESCUCHA ACTIVA",
    "C15": "C15. SABER DECIR (CAPACIDAD EXPRESIÓN)",
    "C16": "C16. CALIDAD DEL TONO",
    "C17": "C17. IMPRESIÓN GENERAL",
    "C18": "C18. CALIDAD DEL ENTORNO",
    "C19": "C19. CALIDAD DE LA RESPUESTA",
    "C20": "C20. PROACTIVIDAD",
    "C21": "C21. SOLUCIÓN AL PRIMER CONTACTO (FCR)",
    "C22": "C22. IMAGEN CORPORATIVA",
    "C23": "C23. DESPEDIDA"
}

# Extraer el código del grupo (C1, C2, ..., C23)
resultado_final['grupo_codigo'] = resultado_final[columna_codigos].str.extract(r'(C\d+)')

# Agregar el nombre del grupo a partir del código
resultado_final['grupo_nombre'] = resultado_final['grupo_codigo'].map(grupo_por_codigo)

# Crear DataFrame final con filas vacías para insertar los nombres de los grupos
df_final = pd.DataFrame(columns=resultado_final.columns)

# Insertar filas con el nombre del grupo
for grupo, df_grupo in resultado_final.groupby('grupo_nombre', sort=False):
    fila_grupo = pd.DataFrame({columna_codigos: [grupo]})
    df_final = pd.concat([df_final, fila_grupo, df_grupo], ignore_index=True)

# Si querés eliminar las columnas auxiliares
df_final = df_final.drop(columns=['grupo_codigo', 'grupo_nombre'])

# Mostrar resultado
print(df_final.head(20))  # podés ajustar este número
```

Se agregan al df las competencias y PE

```python
# Crear un diccionario para mapear la variable a PE y Competencia
variable_to_pe_competencia = {}

# Asumimos que el nuevo archivo tiene las columnas: 'Subgrupo', 'Variables', 'PE', 'Competencia'
for _, row in PE.iterrows():
    variable_to_pe_competencia[row['variables']] = {'PE': row['PE'], 'Competencia': row['Competencia']}

# Ahora, vamos a añadir las columnas 'PE' y 'Competencia' a df_final
# Para esto, necesitamos recorrer df_final y asignar 'PE' y 'Competencia' según corresponda a la variable

df_final['PE'] = df_final['Subgrupo'].map(lambda var: variable_to_pe_competencia.get(var, {}).get('PE', None))
df_final['Competencia'] = df_final['Subgrupo'].map(lambda var: variable_to_pe_competencia.get(var, {}).get('Competencia', None))

# Mostramos el resultado final
print(df_final)
```

```python
# Diccionario para almacenar los resultados
resultados = {}

# Columnas principales a analizar
columnas = [f'C{i}' for i in range(1, 20)]  # C1 a C19

# Recorremos todas las columnas
for columna in columnas:
    if columna in df2.columns:
        # Buscar subgrupos relacionados con esta columna
        subgrupos = [col for col in df2.columns if col.startswith(f'subgrupo_{columna}')]

        for subgrupo in subgrupos:
            # Filtrar solo filas donde la variable principal es 0 o 1
            filas_validas = df2[(df2[columna] == 0) | (df2[columna] == 1)]

            # Contar los 1 en el subgrupo
            cuenta_1 = (filas_validas[subgrupo] == 1).sum()

            # Guardar el resultado (una única vez por subgrupo)
            resultados[subgrupo] = cuenta_1

# Convertimos directamente el diccionario a un DataFrame
resultados_df2 = pd.DataFrame(list(resultados.items()), columns=['Subgrupo', 'Penalizaciones'])

# Mostramos el DataFrame
print(resultados_df2)
```

```python
# Diccionario para almacenar los resultados
resultados = {}

# Columnas principales a analizar
columnas = [f'C{i}' for i in range(1, 18)]  # C1 a C18

# Recorremos todas las columnas
for columna in columnas:
    if columna in df3.columns:
        # Buscar subgrupos relacionados con esta columna
        subgrupos = [col for col in df3.columns if col.startswith(f'subgrupo_{columna}')]

        for subgrupo in subgrupos:
            # Filtrar solo filas donde la variable principal es 0 o 1
            filas_validas = df3[(df3[columna] == 0) | (df3[columna] == 1)]

            # Contar los 1 en el subgrupo
            cuenta_1 = (filas_validas[subgrupo] == 1).sum()

            # Guardar el resultado (una única vez por subgrupo)
            resultados[subgrupo] = cuenta_1

# Convertimos directamente el diccionario a un DataFrame
resultados_df3 = pd.DataFrame(list(resultados.items()), columns=['Subgrupo', 'Penalizaciones'])

# Mostramos el DataFrame
print(resultados_df3)
```

```python
subgrupos_c13 = [col for col in df3.columns if col.startswith('subgrupo_C13')]

penalizaciones_c13 = {}

if 'C13' in df3.columns:
    filas_c13 = df3[(df3['C13'] == 0) | (df3['C13'] == 2)]

    for subgrupo in subgrupos_c13:
        cuenta = (filas_c13[subgrupo] == 1).sum()
        penalizaciones_c13[subgrupo] = cuenta

penalizaciones_c13_chat = pd.DataFrame(
    list(penalizaciones_c13.items()),
    columns=['Subgrupo', 'Penalizaciones']
)

print(penalizaciones_c13_chat)
```

```python
# Convertimos los DataFrames a diccionarios para combinar fácilmente
penalizaciones_generales = dict(resultados_df3.values)
penalizaciones_especiales = dict(penalizaciones_c13_chat.values)

# Actualizamos los valores generales con los valores especiales para los subgrupos C19
penalizaciones_generales.update(penalizaciones_especiales)

# Creamos el DataFrame final combinado
resultado_final3 = pd.DataFrame(
    list(penalizaciones_generales.items()),
    columns=['Subgrupo', 'Penalizaciones']
)

# Mostramos el resultado
resultado_final3
```

```python
# Diccionario para almacenar los resultados
resultados = {}

# Columnas principales a analizar
columnas = [f'C{i}' for i in range(1, 17)]  # C1 a C17

# Recorremos todas las columnas
for columna in columnas:
    if columna in df4.columns:
        # Buscar subgrupos relacionados con esta columna
        subgrupos = [col for col in df4.columns if col.startswith(f'subgrupo_{columna}')]

        for subgrupo in subgrupos:
            # Filtrar solo filas donde la variable principal es 0 o 1
            filas_validas = df4[(df4[columna] == 0) | (df4[columna] == 1)]

            # Contar los 1 en el subgrupo
            cuenta_1 = (filas_validas[subgrupo] == 1).sum()

            # Guardar el resultado (una única vez por subgrupo)
            resultados[subgrupo] = cuenta_1

# Convertimos directamente el diccionario a un DataFrame
resultados_df4 = pd.DataFrame(list(resultados.items()), columns=['Subgrupo', 'Penalizaciones'])

# Mostramos el DataFrame
print(resultados_df4)
```

```python
subgrupos_c13 = [col for col in df4.columns if col.startswith('subgrupo_C13')]

penalizaciones_c13_0_2 = {}

if 'C13' in df4.columns:
    filas_c13_0_2 = df4[(df4['C13'] == 0) | (df4['C13'] == 2)]

    for subgrupo in subgrupos_c13:
        cuenta = (filas_c13_0_2[subgrupo] == 1).sum()
        penalizaciones_c13_0_2[subgrupo] = cuenta

penalizaciones_c13_df = pd.DataFrame(
    list(penalizaciones_c13_0_2.items()),
    columns=['Subgrupo_C13', 'Penalizaciones_C13_0_2']
)

print(penalizaciones_c13_df)

```

```python
# Convertimos los DataFrames a diccionarios para combinar fácilmente
penalizaciones_generales = dict(resultados_df4.values)
penalizaciones_especiales = dict(penalizaciones_c13_df.values)

# Actualizamos los valores generales con los valores especiales para los subgrupos C19
penalizaciones_generales.update(penalizaciones_especiales)

# Creamos el DataFrame final combinado
resultado_final4 = pd.DataFrame(
    list(penalizaciones_generales.items()),
    columns=['Subgrupo', 'Penalizaciones']
)

# Mostramos el resultado
resultado_final4
```
