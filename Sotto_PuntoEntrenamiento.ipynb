# **Ejecutar para SERVICIO TELÉFONICO**

```python
import pandas as pd

# Ruta del archivo Excel en tu Google Drive
file_path = '/content/drive/MyDrive/Sottotempo/Telefonico.xlsx'

# Cargar el archivo Excel
df = pd.read_excel(file_path)

df.head()
```

```python
# Diccionario para almacenar los resultados
resultados = {}

# Columnas principales a analizar
columnas = [f'C{i}' for i in range(1, 24)]  # C1 a C23

# Recorremos todas las columnas
for columna in columnas:
    if columna in df.columns:
        # Buscar subgrupos relacionados con esta columna
        subgrupos = [col for col in df.columns if col.startswith(f'subgrupo_{columna}')]

        for subgrupo in subgrupos:
            # Filtrar solo filas donde la variable principal es 0 o 1
            filas_validas = df[(df[columna] == 0) | (df[columna] == 1)]

            # Contar los 1 en el subgrupo
            cuenta_1 = (filas_validas[subgrupo] == 1).sum()

            # Guardar el resultado (una única vez por subgrupo)
            resultados[subgrupo] = cuenta_1

# Convertimos directamente el diccionario a un DataFrame
resultados_df = pd.DataFrame(list(resultados.items()), columns=['Subgrupo', 'Penalizaciones'])

# Mostramos el DataFrame
print(resultados_df)
```

```python
# Guardamos el DataFrame 'resultados_df' con los nombres reales en un archivo Excel
output_path = '/content/resultados_df_telefonico.xlsx'
resultados_df.to_excel(output_path, index=False)

# Descargar el archivo generado
from google.colab import files
files.download(output_path)
```
