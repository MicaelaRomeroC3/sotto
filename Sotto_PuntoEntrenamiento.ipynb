# **Ejecutar para SERVICIO TELÉFONICO**

```python
import pandas as pd

# Ruta del archivo Excel en tu Google Drive
file_path = '/content/drive/MyDrive/Sottotempo/PE.xlsx'

# Cargar el archivo Excel
df = pd.read_excel(file_path)

df.head()
```

```python
# Diccionario para almacenar los resultados
resultados = {}

# Columnas principales a analizar
columnas = [f'C{i}' for i in range(1, 24)]  # C1 a C23

# Recorremos todas las columnas
for columna in columnas:
    if columna in df.columns:
        # Buscar subgrupos relacionados con esta columna
        subgrupos = [col for col in df.columns if col.startswith(f'subgrupo_{columna}')]

        for subgrupo in subgrupos:
            # Filtrar solo filas donde la variable principal es 0 o 1
            filas_validas = df[(df[columna] == 0) | (df[columna] == 1)]

            # Contar los 1 en el subgrupo
            cuenta_1 = (filas_validas[subgrupo] == 1).sum()

            # Guardar el resultado (una única vez por subgrupo)
            resultados[subgrupo] = cuenta_1

# Convertimos directamente el diccionario a un DataFrame
resultados_df = pd.DataFrame(list(resultados.items()), columns=['Subgrupo', 'Cuenta_1'])

# Mostramos el DataFrame
print(resultados_df)
```

```python
# Diccionario para almacenar los resultados
resultados = {}

# Aquí asignamos las columnas que deseas analizar (por ejemplo, C1, C2, ..., C23)
columnas = [f'C{i}' for i in range(1, 24)]  # C1 a C23

# Recorremos todas las columnas
for columna in columnas:
    # Verificar si la columna existe en el DataFrame
    if columna in df.columns:
        # Buscar los subgrupos relacionados con esta columna
        subgrupos = [col for col in df.columns if col.startswith(f'subgrupo_{columna}')]

        # Para cada subgrupo, contar cuántos '1' hay, pero solo si la variable es 0 o 1
        for subgrupo in subgrupos:
            # Filtrar solo las filas donde la variable asociada (columna) es 0 o 1
            filas_validas = df[(df[columna] == 0) | (df[columna] == 1)]

            # Contar cuántos '1' hay en el subgrupo para las filas válidas
            cuenta_1 = (filas_validas[subgrupo] == 1).sum()  # Contamos solo los 1 en el subgrupo

            # Si no hay 1 en el subgrupo, asignamos 0
            if cuenta_1 == 0:
                cuenta_1 = 0

            # Guardamos el resultado en el diccionario
            if subgrupo not in resultados:
                resultados[subgrupo] = []

            # Añadimos el resultado, incluyendo el nombre del subgrupo
            resultados[subgrupo].append(cuenta_1)

# Convertir los resultados a un DataFrame para exportarlos a Excel
# Asignamos una columna para el nombre del subgrupo para tener más claridad
resultados_con_nombre = []
for subgrupo, cuenta in resultados.items():
    for c in cuenta:
        resultados_con_nombre.append([subgrupo, c])

# Creamos un DataFrame con subgrupo y la cuenta de 1's
resultados_df = pd.DataFrame(resultados_con_nombre, columns=['Subgrupo', 'Cuenta_1'])

# Mostramos el DataFrame
print(resultados_df)
```

```python
# Guardamos el DataFrame 'resultados_df' con los nombres reales en un archivo Excel
output_path = '/content/resultados_df_final.xlsx'
resultados_df.to_excel(output_path, index=False)

# Descargar el archivo generado
from google.colab import files
files.download(output_path)
```
